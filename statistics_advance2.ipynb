1. What is hypothesis testing in statistics?
>> Hypothesis testing is a statistical method used to make decisions or inferences about a population based on a sample of data.

2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?
>> In hypothesis testing, the null hypothesis and the alternative hypothesis represent two opposing ideas about a population. The goal of the test is to decide which one is better supported by the sample data.Null hypothesis is a starting assumption and alternate hypothesis is cnatradiction of null hypothesis.

3. What is the significance level in hypothesis testing, and why is it important?
>> It is a threshold used in hypothesis testing to decide whether to reject the null hypothesis. It represents the maximum probability you're willing to accept.

4. What does a P-value represent in hypothesis testing?
>> P value is the probablity value of the event if the null hypothesis is true.


5. How do you interpret the P-value in hypothesis testing?
>> The p-value helps you decide whether your sample data provides enough evidence to reject the null hypothesis. It tells you the probability of observing your results if it is actually true.

6. What are Type 1 and Type 2 errors in hypothesis testing?
>> ChatGPT said:
Type I error means rejecting a true null hypothesis, while Type II error means failing to reject a false one.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
>> A one-tailed test checks for an effect in one specific direction (greater than or less than), while a two-tailed test checks for any difference, in both directions (not equal to).

8.What is the Z-test, and when is it used in hypothesis testing?
>> A Z-test is a statistical test used to determine whether there is a significant difference between sample and population means or between two sample means. It is used when the population standard deviation is known and the sample size is large >30 . Z-tests assume the data is normally distributed. Common applications include testing the difference between means or proportions, especially in quality control or large-scale experiments. The test produces a Z-score, which is compared to a critical value to accept or reject the null hypothesis.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
>> In hypothesis testing, the Z-score tells you how many standard deviations a data point or sample mean is from the population mean under the null hypothesis. It helps assess whether the observed value is statistically unusual. A higher absolute Z-score means the result is farther from the mean and less likely to have occurred by random chance, indicating potential significance.
given by: z= (samplemean-populationmean)/std deviation

10. What is the T-distribution, and when should it be used instead of the normal distribution?
>> The T-distribution is a probability distribution used in statistics when estimating population parameters, especially the mean, with a small sample size and unknown population standard deviation.It is similar in shape to the normal distribution but has heavier tails, allowing for more uncertainty due to smaller samples. As the sample size increases, the t-distribution approaches the normal distribution.

11.  What is the difference between a Z-test and a T-test?
>> The Z-test is used when the population standard deviation is known and the sample size is large (usually > 30). The T-test is used when the population standard deviation is unknown and the sample size is small (usually < 30).

12. What is the T-test, and how is it used in hypothesis testing?
>> A T-test is used in hypothesis testing to compare means when the sample size is small and the population standard deviation is unknown. It helps determine if observed differences are statistically significant. The result, a t-statistic, is compared to a critical value to accept or reject the null hypothesis.

13.  What is the relationship between Z-test and T-test in hypothesis testing?
>> The Z-test and T-test are both used to test hypotheses about means. They are similar in purpose but differ based on sample size and knowledge of population standard deviation. Use a Z-test when the sample size is large and the population standard deviation is known. Use a T-test when the sample size is small and the population standard deviation is unknown. As the sample size increases, the T-distribution approaches the normal distribution, making the results of both tests nearly identical.

14. What is a confidence interval, and how is it used to interpret statistical results?
>> A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (like the mean) with a certain level of confidence.

15. What is the margin of error, and how does it affect the confidence interval?
>> The margin of error is the range above and below a sample estimate that defines a confidence interval. It shows the possible error due to sampling. A larger margin means more uncertainty.

>> The Chi-square distribution is used in statistics to analyze categorical data. It helps determine whether observed frequencies differ significantly from expected frequencies. Common applications include the Chi-square test of independence, which checks if two categorical variables are related, and the goodness-of-fit test, which assesses how well data match a theoretical distribution. It is applicable when data are in the form of counts and the sample size is large. The distribution is positively skewed, and as degrees of freedom increase, it becomes more symmetric. 19. What is the F-distribution, and when is it used in hypothesis testing?
>>  The F-distribution is a right-skewed probability distribution used in hypothesis testing when comparing variances between two or more groups. It is mainly used in ANOVA (Analysis of Variance) to test if the means of three or more groups are significantly different by analyzing the ratio of between-group variance to within-group variance. It's also used in F-tests to compare two population variances. The F-distribution only takes positive values, since variances can't be negative, and its shape depends on two different degrees of freedom â€” one for the numerator and one for the denominator.

20. What is an ANOVA test, and what are its assumptions?
>>ANOVA is a statistical test used to compare the means of three or more groups to see if at least one differs significantly. It works by analyzing the ratio of between-group variance to within-group variance. The main assumptions are independence of observations, normally distributed data in each group, and equal variances across groups.


21. What are the different types of ANOVA tests?

One-Way ANOVA  Compares means of three or more independent groups based on one factor.

Two-Way ANOVA - Compares means with two independent factors, also checking for interaction effects.

Repeated Measures ANOVA - Compares means when the same subjects are measured under different conditions or times

22. What is the F-test, and how does it relate to hypothesis testing?
>> The F-test is a statistical test that compares two variances to determine if they are significantly different. In hypothesis testing, it is often used to test whether multiple population means are equal, as in ANOVA, by examining the ratio of between-group variance to within-group variance. A high F-value suggests greater variation between groups than within groups, which may indicate a statistically significant difference. The F-test relies on the F-distribution, which depends on two sets of degrees of freedom.

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

data = [50, 52, 47, 49, 51, 53, 48, 52, 50, 49]
mu0 = 50
alpha = 0.05

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

z_stat = (mean - mu0) / (std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", round(mean, 2))
print("Sample Size:", n)
print("Z-statistic:", round(z_stat, 2))
print("P-value:", round(p_value, 4))

if p_value < alpha:
    print("Reject the Null Hypothesis (H0)")
else:
    print("Fail to Reject the Null Hypothesis (H0)")
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

data = [50, 52, 47, 49, 51, 53, 48, 52, 50, 49]
confidence_level = 0.95

mean = np.mean(data)
std_err = stats.sem(data)
n = len(data)

h = std_err * stats.t.ppf((1 + confidence_level) / 2, n - 1)
lower, upper = mean - h, mean + h

print("Sample Mean:", round(mean, 2))
print(f"{int(confidence_level*100)}% Confidence Interval: ({round(lower, 2)}, {round(upper, 2)})")
```

```python
#Practical Part - 2
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
# interpret the results.
import numpy as np
from scipy.stats import norm

sample = [52, 50, 49, 51, 53, 52, 48, 50, 51, 52]

population_mean = 50
population_std = 2

sample_mean = np.mean(sample)
n = len(sample)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample mean and population mean.")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)
data = np.random.normal(loc=52, scale=5, size=100)

population_mean = 50
population_std = 5

sample_mean = np.mean(data)
n = len(data)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)

if abs(z_score) > z_critical:
    decision = "Reject H0"
else:
    decision = "Fail to Reject H0"

print(f"Decision: {decision}")

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, label="Standard Normal Distribution")
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label="Rejection Region")
plt.axvline(z_score, color='blue', linestyle='--', label=f"Z = {z_score:.2f}")
plt.axvline(-z_critical, color='red', linestyle='--', label=f"-Zcrit = {z_critical:.2f}")
plt.axvline(z_critical, color='red', linestyle='--', label=f"Zcrit = {z_critical:.2f}")
plt.title("Two-tailed Z-test Decision Region")
plt.legend()
plt.show()
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05):
    se = sigma / np.sqrt(n)
    z_critical = norm.ppf(1 - alpha/2)
    critical_value_upper = mu0 + z_critical * se
    critical_value_lower = mu0 - z_critical * se
    type1_error = alpha
    beta = norm.cdf((critical_value_upper - mu1) / se) - norm.cdf((critical_value_lower - mu1) / se)

    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10,6))
    plt.plot(x, y0, label="H0 Distribution (Î¼=Î¼0)", color="blue")
    plt.plot(x, y1, label="H1 Distribution (Î¼=Î¼1)", color="red")
    plt.fill_between(x, 0, y0, where=(x <= critical_value_lower) | (x >= critical_value_upper),
                     color="blue", alpha=0.3, label="Type I Error (Î±)")
    plt.fill_between(x, 0, y1, where=(x > critical_value_lower) & (x < critical_value_upper),
                     color="red", alpha=0.3, label="Type II Error (Î²)")
    plt.axvline(critical_value_lower, color="black", linestyle="--")
    plt.axvline(critical_value_upper, color="black", linestyle="--")
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Type I Error (Î±): {type1_error:.3f}")
    print(f"Type II Error (Î²): {beta:.3f}")
    print(f"Power of Test (1 - Î²): {1 - beta:.3f}")


visualize_errors(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.
import numpy as np
from scipy.stats import ttest_ind

def independent_ttest(sample1, sample2, alpha=0.05):
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)  # Welchâ€™s t-test
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Since p-value < {alpha}, we reject the null hypothesis.")
        print("Interpretation: There is a statistically significant difference between the two groups.")
    else:
        print(f"Since p-value >= {alpha}, we fail to reject the null hypothesis.")
        print("Interpretation: There is no statistically significant difference between the two groups.")

np.random.seed(0)
group1 = np.random.normal(loc=50, scale=10, size=30)  # mean=50, std=10
group2 = np.random.normal(loc=55, scale=10, size=30)  # mean=55, std=10

independent_ttest(group1, group2)
```

```python
#7. Perform a paired sample T-test using Python and visualize the comparison results.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_ttest(sample1, sample2, alpha=0.05):
    t_stat, p_value = ttest_rel(sample1, sample2)
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Since p-value < {alpha}, we reject the null hypothesis.")
        print("Interpretation: There is a statistically significant difference between the paired groups.")
    else:
        print(f"Since p-value >= {alpha}, we fail to reject the null hypothesis.")
        print("Interpretation: There is no statistically significant difference between the paired groups.")

    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.boxplot([sample1, sample2], labels=["Sample 1", "Sample 2"])
    plt.title("Boxplot Comparison")
    plt.ylabel("Values")

    plt.subplot(1,2,2)
    for i in range(len(sample1)):
        plt.plot([1,2], [sample1[i], sample2[i]], marker="o", color="gray", alpha=0.6)
    plt.xticks([1,2], ["Sample 1", "Sample 2"])
    plt.title("Paired Differences Plot")

    plt.tight_layout()
    plt.show()

np.random.seed(1)
before = np.random.normal(loc=50, scale=5, size=30)
after = before + np.random.normal(loc=2, scale=5, size=30)

paired_ttest(before, after)
```

```python
#8. Simulate data and perform both Z-test and T-test, then compare the results using Python.
 import numpy as np
from scipy.stats import norm, ttest_1samp

np.random.seed(42)
data = np.random.normal(loc=52, scale=5, size=30)

population_mean = 50
population_std = 5

sample_mean = np.mean(data)
n = len(data)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
z_p_value = 2 * (1 - norm.cdf(abs(z_score)))

t_stat, t_p_value = ttest_1samp(data, population_mean)

print("Z-test Results")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {z_p_value:.4f}\n")

print("T-test Results")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {t_p_value:.4f}\n")

alpha = 0.05
print("Decision at Î± = 0.05")
print(f"Z-test: {'Reject H0' if z_p_value < alpha else 'Fail to Reject H0'}")
print(f"T-test: {'Reject H0' if t_p_value < alpha else 'Fail to Reject H0'}")
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_crit * std_err
    return mean - margin, mean + margin

np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=30)
ci = confidence_interval(sample, confidence=0.95)

print(f"Sample Mean: {np.mean(sample):.2f}")
print(f"95% Confidence Interval: {ci}")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_crit * std_err
    return mean - margin, mean + margin

np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=30)
ci = confidence_interval(sample, confidence=0.95)

print(f"Sample Mean: {np.mean(sample):.2f}")
print(f"95% Confidence Interval: {ci}")
'''
A confidence interval (CI) gives a range of values within which the true population mean is likely to fall, based on the sample data.

For example, if the 95% CI is (48.2, 52.7), it means we are 95% confident that the true population mean lies between 48.2 and 52.7.

It doesnâ€™t mean thereâ€™s a 95% probability that the mean is in that range (the true mean is fixed), but rather that 95% of such intervals computed from repeated samples would contain the true mean.

'''
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior, sensitivity, false_positive, evidence_positive):
    posterior = (sensitivity * prior) / evidence_positive
    return posterior

prior = 0.01
sensitivity = 0.95
false_positive = 0.05
no_disease = 1 - prior

evidence_positive = (sensitivity * prior) + (false_positive * no_disease)

posterior = bayes_theorem(prior, sensitivity, false_positive, evidence_positive)

print(f"Prior Probability of Disease: {prior:.2f}")
print(f"Posterior Probability of Disease given Positive Test: {posterior:.4f}")
'''Even though the test is highly sensitive (95%), the posterior probability is only about 16% because the disease is rare.

This shows how Bayesian inference balances prior beliefs with new evidence.'''
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

data = pd.DataFrame({
    "Gender": ["Male", "Male", "Male", "Female", "Female", "Female", "Male", "Female", "Male", "Female"],
    "Preference": ["A", "B", "A", "B", "A", "B", "B", "A", "A", "B"]
})

contingency_table = pd.crosstab(data["Gender"], data["Preference"])
print("Contingency Table:")
print(contingency_table)

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"\nChi-square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")
print("\nExpected Frequencies:")
print(expected)

alpha = 0.05
if p < alpha:
    print("\nReject the null hypothesis: Gender and Preference are dependent.")
else:
    print("\nFail to reject the null hypothesis: Gender and Preference are independent.")
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

observed = np.array([8, 10, 9, 12, 11, 10])
expected = np.array([10, 10, 10, 10, 10, 10])

chi2, p = chisquare(f_obs=observed, f_exp=expected)

print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected)
print(f"\nChi-square Statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")

alpha = 0.05
if p < alpha:
    print("\nReject the null hypothesis: The die is not fair.")
else:
    print("\nFail to reject the null hypothesis: The die appears fair.")
```
